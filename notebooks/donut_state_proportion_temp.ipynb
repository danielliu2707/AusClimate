# Transformation of Rainfall data for radial chart:

```python
import pandas as pd
import numpy as np
import os
```

```python
state_proportion_temp = pd.read_csv("../data/state_temp.csv")
state_proportion_temp.head()
```

```python
state_proportion_temp
```

```python
# Group by Month and State, then sum the TemperatureMean
monthly_state_totals = state_proportion_temp.groupby(['Month', 'officialnamestate'])['TemperatureMean'].sum().reset_index()

# Group by Month to get the total TemperatureMean for each month
monthly_totals = monthly_state_totals.groupby('Month')['TemperatureMean'].sum().reset_index()
monthly_totals.rename(columns={'TemperatureMean': 'TotalTemperature'}, inplace=True)

# Merge to get the total temperature in each month for each state
state_proportion_temp = pd.merge(monthly_state_totals, monthly_totals, on='Month')

# Calculate the proportional contribution
state_proportion_temp['ProportionalContribution'] = state_proportion_temp['TemperatureMean'] / state_proportion_temp['TotalTemperature'] * 100

# Selecting relevant columns
state_proportion_temp = state_proportion_temp[['Month', 'officialnamestate', 'ProportionalContribution']].rename(columns={'ProportionalContribution': "ProportionalTemp"})

state_proportion_temp
```

```python
# Exporting data:
state_proportion_temp.to_csv("../data/state_proportional_temp.csv")
```
